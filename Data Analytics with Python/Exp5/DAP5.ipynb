## UPI Transactions Dataset (Sampling Distribution)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r'MyTransaction.csv')
df
```

```python
num_samples = 30
sample_size = 75
samples = []
for i in range(num_samples):
    sample = df['Balance'].sample(n=sample_size, replace=False)
    samples.append(sample)
```

```python
samples
```

```python
sample_means = []
for sample in samples:
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)
for i in range(num_samples):
    print(f"Sample {i+1}: Mean = {sample_means[i]}")
```
